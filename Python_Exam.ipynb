**1) Fibonnacci  Sequence**

```python
def fibonacci(n:int)->int: # This function receives an int value and returns an int 
    if n<2:
        return n # If n es less than 2 , by definition the return is n
    else:
        return fibonacci(n-1)+fibonacci(n-2) # If n>=2 the code is recursive until reach 0

fibo = int(input("Which Fibonnacci position do you want? "))
print(f"The {fibo} position is {fibonacci(fibo)}")
```

```python
def fibonnacciUpTo(n: int)->list: #This function receives an int and returns a list
    answer = []
    for i in range(n):
        answer.append(fibonacci(i)) # This parts stores the fibonnacci values returned by the function above
    return answer

limit = int(input("How many Fibonnacci numbers do you want? "))
print(f"{fibonnacciUpTo(limit)}")
```

**2) Monte Carlo Simulation**

```python
import random
import matplotlib.pyplot as plt
import numpy as np

n = int(input("How many points do you want to make: "))
l_square = 50 # this is the size of the rectangle
radius = l_square/2  # this is the radius of the circle
in_circle = 0 #This variable counts how many points will be in the circle
x_coord, y_coord = [],[] #This lists will be used in the plot to store the points
for i in range(n):
    x = random.uniform(-radius,radius) 
    y = random.uniform(-radius,radius) #This part generates random numbers
    x_coord.append(x)
    y_coord.append(y)
    if (x**2 + y**2) <= radius**2: #This part checks if the random point (x,y) is inside the triangle
        in_circle +=1
pi_estimate = 4*in_circle/n #This variable shows the estimate value
print(f"The estimate value of pi is {pi_estimate}")

#This is the circles's parametric form in (0,0) center, 
#with linspace we create 100 angles to be store and plot
xc = 0 + radius*np.cos(np.linspace(0,2*np.pi,100))
yc = 0 + radius*np.sin(np.linspace(0,2*np.pi,100)) 

#This part plots the rectangle [width_start,width_end,height_start,height_end]
plt.axis([-l_square/2, l_square/2, -l_square/2, l_square/2]) 
plt.plot(xc,yc,color="purple") #This part plot the circle
plt.scatter(x_coord,y_coord) ##This part plot the points
plt.show()
```

**3) Caesar Cipher**

```python
import string
alphaphet = string.ascii_lowercase #This import the alphaphet in lower case
def cipher(name:str,n:int)->str: #This function receives an int and a string and return a string
    new_name = "" #This variable will be store the new name
    for i in range(len(name)):
        if name[i] == " ": #If the name has white space, the cipher will put a white space
            s = " "
            new_name += s
        elif name[i] not in alphaphet: #If there is a rare character, the iteration pass
            pass
        else:
            s = alphaphet.index(name[i]) #this part identify the position in the word
            new_name += alphaphet[s+n] #This part is creating the new word with the shift
    return new_name

name = input("Introduce the name that will be encrypted: ").lower()
n = int(input("Introduce the shift: "))
print(f"Your encrypted new name is: {cipher(name,n)}\nYour shift key is {n} ")
```

**4) List of Lists**

```python
def list_div(l:list,ndiv:int)->list:
    '''
    This function receives a list and a int an return a list
    '''
    
    #This create the new list using the ndiv number, the range divides the 
    #original list with the third parameter
    
    list_group = [l[i:i+ndiv] for i in range(0,len(l),ndiv)] 
    return list_group

nele = int(input("How many values do you want in your list?: "))
ndiv = int(input("How many divisions do you want?: "))
l = [i for i in range(nele)] #This creates the elements in the list
print(f"Here is your new list\n{list_div(l,ndiv)}")
```

**6) Hangman Game**

```python
data_2 = open("contries.txt", "r")
content_2 = data_2.read()
content_list_2 = content_2.split("\n") #Storing the data
data_2.close()
```

```python
import random
word = random.choice(content_list_2).upper() #This part stores in upper case all countries
tries = 5
finish = False
guess_word = [] #This variable store the correct attempts
print(f"The word has {len(word)} letters") #This part shows the letter
aux_show = 0
for i in word: #This for loop creates a clue if the aux_show is different from 2
    if aux_show!=2:
        print("_", end=" ")
    else:
        print(i, end=" ")
        guess_word.append(i) #We store the clue
    aux_show +=1
print("")
while not finish and tries>0:
    try_game = input(f"You have {tries} tries left, digit your guess: ").upper() #Store the user's try
    if try_game not in word: #If the try word is not in the original word the number of tries decrease
        tries -= 1
        print("Wrong! Dx")
    else:
        print("Good!")
        guess_word.append(try_game) #If the try word is in the original word, we add the word in the correct attempts
        for i in word:
            if i in guess_word:
                print(i, end=" ") #This part shows the letters that had been found
            else:
                print("_", end=" ") #This part shows the missing letters
    print("")
    
    finish = True #We change the variable to True, if the for loop does not change its value, the game finish
    for i in word:
        if i not in guess_word: #This part checks if all the letters are complete
            finish = False

if finish:
    print("You win")
else:
    print("You lose")
    print(f"The word was {word}")
    
```

**7) Pascalâs Triangle**

```python
n = int(input("Digit the number of rows: "))
ans = []
for i in range(n):
    pascal_row= [] #This variable stores the triangle rows
    for j in range(i+1):
        if j==0 or j==i: #If the j is equal to 0 or j == i by the definition is 1
            pascal_row.append(1)
        else:
            pascal_row.append(ans[i-1][j-1] + ans[i-1][j]) #This sums the bounds for the last iteration
    ans.append(pascal_row) #We sum like this because its a way to find the values for the triangle
    
for i in range(n):
    for j in range(n-i-1): #The n-i-1 creates the horizontal white spaces for the visualization
        print(" ",end = "")
    for j in range(i+1): #This part shows each list with a slight space
        print(ans[i][j],end=" ")
    print("")#This part makes a space between each line
    
```
